```python
x_value_01 = np.arange(1, 10)
y_value_01 = 2*x_value_01

plt.plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
plt.bar(x_value_01, y_value_01, color='g', label='bar plot')
plt.title('Hello')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks(ticks=np.arange(1, 13, 1), rotation=75)
plt.legend()
plt.show()
```

```python
import numpy as np

x_value_01 = np.arange(0, 10)
x_value_02 = np.arange(0, 20)
y_value_01 = 2* x_value_01
y_value_02 = 2* x_value_02

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6, 3))

ax[0].plot(x_value_01, y_value_01, color='r', label='red_line', marker='o', linestyle='dashed')
ax[1].bar(x_value_02, y_value_02, color='g', label='bar_plot')

ax[0].set_xlabel('ax[0] x axis')
ax[1].set_xlabel('ax[1] x axis')

ax[0].legend()
ax[1].legend()

ax[0].set_title('Hello line')
ax[1].set_title('Hello bar')

plt.show()
```

```python
import numpy as np
x_value_01 = np.arange(1, 10)
x_value_02 = np.arange(1, 20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6, 6))

ax[0][0].plot(x_value_01, y_value_01, color='red', linestyle='dashed', marker='o', label='red line')
ax[1][0].plot(x_value_01, y_value_01, color='g', linestyle='dashed', marker='o', label='green line')
ax[0][1].bar(x_value_02, y_value_02, color='g', label='green bar')
ax[1][1].bar(x_value_02, y_value_02, color='r', label='red bar')

ax[0][0].set_xlabel('ax[0][0] x axis')
ax[1][0].set_xlabel('ax[1][0] x axis')
ax[0][1].set_xlabel('ax[0][1] x axis')
ax[1][1].set_xlabel('ax[1][1] x axis')

for i in range(2):
    for j in range(2):
        ax[i][j].legend()

plt.show()
```
